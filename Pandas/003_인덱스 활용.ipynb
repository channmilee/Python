### 3-1. 특정 열을 행 인덱스로 설정

* DataFrame 객체.set_index(['열 이름'] 또는 '열 이름')
  * 기존 객체 변경 X, 새로운 데이터 프레임 객체 반환

```python
import pandas as pd
```

```python
exam_data = {'이름':['서준','우현','인아'],
             '수학':[ 90,80,70],'영어':[ 98,89,95],
             '음악':[ 85,95,100],'체육':[ 100,90,90]}

df = pd.DataFrame(exam_data)
df
```

```python
df.set_index('이름', inplace = True)
df
```

```python
mdf = df.set_index('음악')
mdf
```

```python
df3 = df.set_index(['음악','수학'])
df3
```

### 3-2. 행 인덱스 재배열

* DataFrame 객체.reindex(새로운 인덱스 배열)
  * 기존 객체 변경 X, 새로운 데이터 프레임 객체 반환

```python
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2':[7,8,9], 'c3' : [10,11,12], 'c4':[13,14,15]}

df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
df
```

```python
new_index = ['r0','r1','r2','r3','r4']

ndf = df.reindex(new_index)
ndf
```

```python
new_index = ['r0','r1','r2','r3','r4']

ndf2 = df.reindex(new_index, fill_value = 0)
ndf2
```

### 3-3. 행 인덱스 초기화

* 정수형 위치 인덱스로 초기화 : DataFrame 객체.reset_index() 

```python
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2':[7,8,9], 'c3' : [10,11,12], 'c4':[13,14,15]}

df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
df
```

```python
# 행 인덱스를 정수형으로 초기화
ndf = df.reset_index()
ndf
```

### 3-4. 행 인덱스를 기준으로 데이터프레임 정렬

* 행 인덱스 기준 정렬 : DataFrame 객체.sort_index()

```python
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2':[7,8,9], 'c3' : [10,11,12], 'c4':[13,14,15]}

df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
df
```

```python
ndf = df.sort_index(ascending = False)
ndf
```

### 3-5. 특정 열의 데이터 값을 기준으로 데이터프레임 정렬

* 열 기준 정렬 : DataFrame 객체.sort_values(by = ' ')

```python
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2':[7,8,9], 'c3' : [10,11,12], 'c4':[13,14,15]}

df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
df
```

```python
ndf = df.sort_values(by = 'c1', ascending = False)
ndf
```
