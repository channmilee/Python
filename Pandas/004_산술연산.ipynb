## 시리즈 연산 

### 4-1. 시리즈 vs 연산

* Series 객체 + 연산자 + 숫자

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
print(student1)
```

```python
percentage = student1/200

print(percentage)
print(type(percentage))
```

### 4-2. 시리즈 vs 시리즈

* Series + 연산자 + Series2

```python
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90, '영어':80})

print(student1)
print('\n')
print(student2)
```

```python
add = student1 + student2
sub = student1 - student2
multi = student1 * student2
div = student1 / student2

print(type(div))
print('\n')

result = pd.DataFrame([add,sub,multi, div], index = ['덧셈','뺄셈','곱셈','나눗셈'])
print(result)
```

```python
# NaN 값이 있는 시리즈 연산
student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90})

print(student1)
print('\n')
print(student2)
print('\n')
```

```python
add = student1 + student2
sub = student1 - student2
multi = student1 * student2
div = student1 / student2

print(type(div))
print('\n')

result = pd.DataFrame([add,sub,multi, div], index = ['덧셈','뺄셈','곱셈','나눗셈'])
result
```

### 4-3. 연산 메소드

* Series1.add(Series2, fill_value = 0)

```python
student1 = pd.Series({'국어':np.nan, '영어':80, '수학':90})
student2 = pd.Series({'수학':80, '국어':90})

print(student1)
print('\n')
print(student2)
print('\n')
```

```python
# NaN 값 0으로 대체
sr_add = student1.add(student2, fill_value = 0)
sr_sub = student1.sub(student2, fill_value = 0)
sr_mul = student1.mul(student2, fill_value = 0)
sr_div = student1.div(student2, fill_value = 0)

result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div], index = ['덧셈','뺄셈','곱셈','나눗셈'])

result
```

## 데이터 프레임 연산

### 4-4. 데이터 프레임 vs 숫자

* DataFrame 객체 + 연산자 + 숫자

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]

df.head()
```

```python
print(type(df))
```

```python
addition = df + 10

print(type(addition))
addition.head()
```

### 4-5. 데이터 프레임 vs 데이터 프레임

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]

print(type(df))
df.tail()
```

```python
addition = df + 10

print(type(addition))

addition.tail()
```

```python
subtraction = addition - df

print(type(subtraction))
subtraction.tail()
```
